```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
```

```python
import io
data1 = pd.read_csv(io.BytesIO(uploaded['2018_data.csv']), parse_dates=True)
data2 = pd.read_csv(io.BytesIO(uploaded['2019_data.csv']), parse_dates=True)

```

```python
data1.head()
data2.head()
```

```python
data_2018 = data1.fillna(0, inplace=False)
data_2019 = data2.fillna(0, inplace=False)
```

```python
import matplotlib.pyplot as plt
```

```python
data_2019.describe()
```

```python
def graphs(year1, year2, name):
  from matplotlib import style
  plt_1 = plt.figure(figsize=(12, 5))
  plt.plot(year1, name, data=data_2018, marker='*', markerfacecolor='white', markersize=12, color='cyan',label= year1)
  plt.plot(year2, name, data=data_2019, marker='*', markerfacecolor='white', markersize=12, color='blue', label= year2)
  plt.title(f'{name} for {year1} and {year2}', fontsize=25)
  plt.ylabel(name, fontsize=15)
  plt.xlabel("Date", fontsize=15)
  plt.legend()
  plt.axes().set_facecolor("gray")



```

```python
graphs("2018", '2019', "NO2_ppb")
```

```python
graphs("2018", '2019', "OZONE_ppm")
```
